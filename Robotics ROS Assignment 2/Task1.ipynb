```python
print('Assignment Operations')
print('enter numbers to perform operation')
a = input('Enter a number (a)')  
b = input('Enter a number (b)')  

# Addition of a & b 
sum = float(a) + float(b)
# Subtraction of a & b 
min = float(a) - float(b) 
# Multiplication of a & b 
mul = float(a) * float(b)  
# Division of a & b 
div = float(a) / float(b)
# Modulus of a & b
mod = float(a)%float(b)
# Floor Division a & b
fdiv = float(a)//float(b)
# Exponent of a ^ b
exp = float(a)**float(b)

# Display the Addition of a & b 
print('Addition a+b=',sum)
# Display the Subtraction of a & b 
print('Subtraction a-b=',min)  
# Display the Multiplication of a & b 
print('Multiplication a*b=',mul) 
# Display the Division of a & b 
print('Division a/b=',div)
# Display the Modulus of a & b
print('Modulus a%b=',mod)
# Display the Floor Division a & b
print('Floor division a//b=',fdiv)
# Display the Exponent of a ^ b
print('Exponent a**b=',exp)
```
